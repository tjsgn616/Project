```python
import webbrowser
import folium
import json
import pandas as pd
import numpy as np
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

from google.colab import drive

```

```python
drive.mount("/content/drive")
DATA = "/content/drive/MyDrive/Colab Notebooks/data/mise_of_region_2020.csv"
mise_of_region_2020data = pd.read_csv(DATA)
print(pd.read_csv(DATA))
print(mise_of_region_2020data.columns)
print(mise_of_region_2020data.index)
```

## 데이터 정리
- 지역명 데이터와 미세먼지량 데이터 분류

```python
region = mise_of_region_2020data[ ["month"] ]
region.info()
region.head()
print(region)
print(type(region))
## 컬럼명 바꿀 필요 있음.

mise_of_month_2020data = mise_of_region_2020data
del mise_of_month_2020data['month']
mise_of_month_2020data
```

```python
region.rename(columns={'month':'region'}, inplace = True)
print(region)
```

```python
mean_of_mise= mise_of_month_2020data.mean(axis='columns')
mean_of_mise = pd.DataFrame(mean_of_mise)
print(type(mean_of_mise))

print(mean_of_mise)

mean_of_mise.columns = ['Mean']
print(mean_of_mise)
```

1. 데이터 프레임 + 시리즈 결합
2. 정렬
3. 분리

```python
## 1. 데이터 결합
mean_of_region = pd.concat([region,mean_of_mise], axis=1)

## 2. 평균 미세먼지를 기준으로 내림차순
mean_of_region = mean_of_region.sort_values('Mean', ascending=False)
print(mean_of_region.columns)
# ## 3. 지역과 평균미세먼지 분류
# sec_region = mean_of_region[ [ ]]
mean_of_region.rename(columns={'month':'month', '0':'Mean'}, inplace=True)
#mean_of_region.rename(columns={'month','Mean'}, inplace=True)
print(mean_of_region)
```

mean_of_region['month'] : 지역구분


mean_of_region[0] : 미먼 평균 수치




```python
mean_of_region.loc[:, 'Eng'] = pd.Series(['Gyeonggi-do', 'Chungcheongnam-do', 'Sejongsi','Chungcheongbuk-do','Seoul','Daegu','Incheon','Jeollabuk-do','Daejeon','Gyeongsangbuk-do','Ulsan','Gangwon-do','Jeju-do','Busan','Gwangju','Gyeongsangnam-do','Jeollanam-do'], index=mean_of_region.index)
mean_of_region
```

```python
# def folium_del_legend(choropleth: folium.Choropleth):
#   del_list = []
#   for child in choropleth._children:
#     if child.startswith('color_map'):
#       del_list.append(child)
#   for del_item in del_list:
#     choropleth._children.pop(del_item)
#   return choropleth

center = folium.Map(location=[36,128], zoom_start=7.4, tiles='openStreetMap')
# 출처 : https://github.com/southkorea/southkorea-maps/
with open('/content/drive/MyDrive/Colab Notebooks/data/skorea-provinces-2018-geo.json', mode='rt', encoding='utf-8') as f:
  geo = json.loads(f.read())
  f.close()
folium.GeoJson(geo, name='seoul_provinces').add_to(center)

# folium_del_legend(
folium.Choropleth(
    geo_data = geo,
    data = mean_of_region,
    columns = ['Eng', 'Mean'],
    fill_color='OrRd',
    key_on='feature.properties.name_eng',
    legend_name = '미세먼지 농도'
    
).add_to(center)
center.save('/content/drive/MyDrive/folium_kr.html')
#webbrowser.open_new("folium_kr.html")
```
